```python
import cv2 
  
# Loading image
image = cv2.imread('C://image1/Dolph.jpg')

cv2.imshow('Original Image', image)
cv2.waitKey()

img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('C://image1/dolphGS.jpg', img)

cv2.imshow('Grayscale Image', img)
cv2.waitKey()

cv2.destroyAllWindows() 
```

```python
from skimage import io
import matplotlib.pyplot as plt

img = io.imread('C://image1/dolphGS.jpg')
axis = plt.hist(img.ravel(), bins = 256)
plt.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt

image = io.imread('C://image1/Dolph.jpg')

plt.hist(image[:, :, 0].ravel(), bins = 256, color = 'red', alpha = 0.5)
plt.hist(image[:, :, 1].ravel(), bins = 256, color = 'green', alpha = 0.5)
plt.hist(image[:, :, 2].ravel(), bins = 256, color = 'blue', alpha = 0.5)
plt.xlabel('Intensity')
plt.ylabel('Total Count')
plt.show()
```
