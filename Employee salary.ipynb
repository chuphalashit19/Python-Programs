```python
n = int(input("Enter No. of employees:"))

for i in range(n):
    salary = int(input("Enter the basic salary of employee (" +str(i+1)+ "):"))
    emp_class = int(input("Enter the class of employee (" +str(i+1)+ "):"))
    exp = int(input("Enter experience years for employee (" +str(i+1)+ "):"))
    if salary < 200:    
        allowance = 0.07 * salary
    elif salary in range(200, 400):
        allowance = 0.10 * salary
    elif salary >= 400:
        allowance = 0.15 * salary
        
    if exp < 10:
        tech_allowance = 0.12 * salary
    elif exp >= 10:
        tech_allowance = 0.2 * salary
    
    gross_salary = salary + allowance + tech_allowance
    
    if emp_class == 1:
        if gross_salary < 300:
            retirement_deduction = 0.02 * gross_salary
        elif gross_salary in range(300, 500):
            retirement_deduction = 0.04 * gross_salary
        elif gross_salary >=  500:
            retirement_deduction = 0.06 * gross_salary
        
    elif emp_class == 2:
        if gross_salary < 200:
            retirement_deduction = 0.05 * gross_salary
        elif gross_salary in range(200, 400):
            retirement_deduction = 0.07 * gross_salary
        elif gross_salary in range(400, 600):
            retirement_deduction = 0.09 * gross_salary
        elif gross_salary >= 600:
            retirement_deduction = 0.12 * gross_salary
    
    elif emp_class == 3:
        retirement_deduction = 0.00
        
    net_salary = gross_salary - retirement_deduction
    
    print("-------------------- emp(" +str(i+1)+ ") -------------------------------")
    print("Basic Salary  =  " +str(salary)+ "    CLASS   =   " +str(emp_class))
    print("Experience Years  =  " +str(exp)+ "   Basic Allowance = " +str(round(allowance,2)))
    print("Technical Allowance = " +str(round(tech_allowance, 2))+ "   Gross salary = " +str(round(allowance, 2)))
    print("Retirement Deduction = " +str(round(retirement_deduction, 2))+ "   Net Salary = " +str(round(net_salary, 2)))
    print("------------------------------------------------------------------------")
    print()
    print()
```
