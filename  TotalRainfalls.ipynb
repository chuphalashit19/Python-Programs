```python
print("--------Rainfalls for 2020--------")

rainf = []
months = ["January", "February", "March", "April", "May", "June", "July",
          "August", "September", "October", "Novemeber", "December"]

total = 0

for i in range(12):
    rainfall = int(input("Rainfall for Month" +str(i+1)+ ": "))         #taking input from user for each rainfall
    total += rainfall                                                   #adding the each rainfall to total
    rainf.append(rainfall)                                              #appeding the rainfalls to list[] rainf

print()
print("--------Months Rainfalls for 2020--------")                      #printing months rainfalls line by line
for i in range(12):
    print(months[i], ":" ,rainf[i])

print("Total Rainfall amount for the year:", total)                    #printing average rainfall for each  month
print("\n------Average Monthly Rainfall------")
for i in range(12):
    print(months[i], ":" , round(rainf[i]/12, 2))
maximum_rainfall = 0 
mx = 0
minimum_rainfall = 1000000
mn = 0

# loop to find the maximum rainfall and the month it happened
for i in range(12):
    if maximum_rainfall < rainf[i]:
        maximum_rainfall = rainf[i]
        mx = i
    
# loop to find the minimum rainfall and the month it happened
for i in range(12):
    if minimum_rainfall > rainf[i]:
        minimum_rainfall = rainf[i]
        mn = i
        
print()
print("Highest Rainfall Amount = ", months[mx][:3:], ":", maximum_rainfall)
print("Lowest Rainfall Amount = ", months[mn][:3:], ":", minimum_rainfall)
```
