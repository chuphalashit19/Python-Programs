```python
#Possible Absolute Mean

import numpy as np

import pandas as pd

from sklearn.linear_model import LinearRegression

  

df = pd.read_csv(r'C:\Users\Dedrick\Documents\650_project\Community Data.csv')

df.head()

df = df.replace({'?': np.NaN})


x = np.array(df['ViolentCrimesPerPop']).reshape((-1,1))

y =np.array(df['agePct16t24'])

model = LinearRegression()

model.fit(x,y)

model = LinearRegression().fit(x,y)

x_pred=model.predict(x)

print('Predicted Response:', x_pred, sep='\n')

def mean_absolute_percentage_error(x_true, x_pred):

    x_true, x_pred = np.array(x_true), np.array(x_pred)

    return np.mean(np.abs((x_true - x_pred)/ x_true))*(100)

print (mean_absolute_percentage_error)
```
