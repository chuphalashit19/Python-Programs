```python
lis = list(map(int, input("Enter the numbers for the list: ").split()))
while 1:
    length = total = avg = 0
    maximum = -1
    minimum = 10000000000
    print()
    print("1. Count of the numbers in the list.")
    print("2. Sum of the numbers in the list.")
    print("3. Average of the numbers in the list.")
    print("4. Maximum of the numbers in the list.")
    print("5. Minimum of the numbers in the list.")
    print("6. Otherwise to exit.")
    print("Enter you choice: ")
    ch = int(input())
    if ch == 1:
        length = len(lis)
        print("Count of the numbers in the list:", length)

    elif ch == 2:
        for i in lis:
            total += i
        print("Sum of the numbers in the list:", total)

    elif ch == 3:
        for i in lis:
            total += i
        avg = total/len(lis)
        print("Average of the numbers in list:", round(avg, 2))

    elif ch == 4:
        for i in lis:
            if maximum < i:
                maximum = i
        print("Maximum number in list:", maximum)

    elif ch == 5:
        for i in lis:
            if minimum > i:
                minimum = i
        print("Minimum number in list:", minimum)
    
    else:
        print("\n----Exitinng------\n")
        break
```
