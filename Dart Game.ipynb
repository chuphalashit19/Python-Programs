```python
import math

def diff(a):
    print("Hit Point is: ", a)
    print("Centre is:  (0, 0)")
    d = math.sqrt((pow(a[0], 2)-0)+(pow(a[1], 2)-0))
    print("The distance is: ", d)
    if (d < 11):
        print("Result: True")
        print("Hit the board!")
    else:
        print("Result: false")
        print("Outside of the dart board!")
    
    if (0 < d <= 3):
        print("Score: 10")
    elif (3 < d <= 7):
        print("Score: 5")
    elif (7 < d <= 11):
        print("Score: 3")
    else:
        pass
    print("-----------------------------------------------")
    
dart = [(1, 5), (2, 6), (3, 9), (6, 3), (7, 2), (8, 8), (3, 6), (4, 4), (9, 9), (5, 8)]
r = 11

print("               ^^** Playing Dart Game **^^               ")
print("*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^*^")
print("--------------------------------------------------------------")
print(list(map(diff, dart)))
```
