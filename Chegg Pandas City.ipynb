```python
import pandas as pd

# used to print the data in csv file
pd.read_csv("city.csv")

#storing data of csv file in variable df
df = pd.read_csv("city.csv")

# city with maximum population
max_pop = df.max()["city_pop"] # max popultion in column city_pop

#looping through the cities and population
for index, row in df.iterrows(): 
    pop = row["city_pop"]
    if max_pop == pop:
        print(row['city'])

#prints city with maximum population

# the place where Jeff Elliott lives
for index, row in df.iterrows():
    first, last = row['first'], row['last']
    if first == "Jeff" and last == "Elliott":
        print(row['street'],",",row['city'],",",row['state'])

# the top two most common last name
l = []
for index, row in df.iterrows():
    # creating a list with last names
    last = row['last']
    l.append(last)
    
# using counter to find the count of each last name and then displaying two most common ones
from collections import Counter
print(Counter(l).most_common(2))
```
