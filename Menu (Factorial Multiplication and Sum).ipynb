```python
while 1:
    n = int(input("Enter any integer value:"))
    print("    1. Factorial of " +str(n))
    print("    2. Sum of values from 1 to " +str(n))
    print("    3. Multiplication table of " +str(n))
    print("    4. Exit")
    ch = int(input("Enter your choice:"))
    print()
    
    if ch == 1:
        fact = 1
        if n == 0:
            fact = 1
        else:
            for i in range(1, n+1):
                fact = fact*i
                
        print("Factorial of (" +str(n)+ ") =", fact)
        print()
        
    elif ch == 2:
        total = 0
        for i in range(1, n+1):
            total += i
        print("Sum of values from 1 to " +str(n)+ " =", total)
        print()
        
    elif ch == 3:
        print("Multiplication table for (" +str(n)+ ")")
        print("--------------------------------")
        for i in range(1, 13):
            print(str(i)+ "*" +(str(n))+ "=" +str(i*n))
        print("--------------------------------")
        print()
    
    else:
        print("Exiting!!!!")
        break
```
