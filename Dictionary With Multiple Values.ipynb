```python
empdict = {}

empid = [44, 45, 46, 47, 48, 49, 50, 51, 52]
fname = ["Simon", "Thomas", "Eugene", "Andrew", "Ruth", "Barry", "Sidney", "Jeffrey", "Doris"]
lname = ["Martinez", "Goldstein", "Comelsen", "Petculescu", "Stadick", "Scardelis", "Hunter", "Evans", "Berndt"]
nin = ["HH 45 09 73 D", "SA 75 35 42 B", "NE 22 63 82", "XY 29 87 61 A", "MA 12 89 36 A", "AT 20 73 18", "HW 12 94 21 C", "LX 13 26 39 B", "YA 49 88 11 A"]
deptId = [1, 2, 2, 1, 15, 2, 6, 6, 3]

empdict["employee_id"] = empid
empdict["first_name"] = fname
empdict["last_name"] = lname
empdict["nin"] = nin
empdict["department_id"] = deptId

k = 0
keys = list(empdict)

for i in range(len(empdict[keys[0]])):
    print("Employee ID:", empdict[keys[0]][k])
    print("First Name:", empdict[keys[1]][k])
    print("Last Name:", empdict[keys[2]][k])
    print("NIN:", empdict[keys[3]][k])
    print("Department_ID:", empdict[keys[4]][k])
    k += 1
    print()
```

```python
name = input("Enter the name for which you wish to delete details: ")
for i in empdict:
    if i == "first_name":
        for j in range(len(empdict[i])):
            if name == empdict[i][j]:
                for k in empdict:
                    del empdict[k][j]
                break

k = 0
keys = list(empdict)

for i in range(len(empdict[keys[0]])):
    print("Employee ID:", empdict[keys[0]][k])
    print("First Name:", empdict[keys[1]][k])
    print("Last Name:", empdict[keys[2]][k])
    print("NIN:", empdict[keys[3]][k])
    print("Department_ID:", empdict[keys[4]][k])
    k += 1
    print()
```
