```python
# list.py

lis = ["Dressing", "oxygen admin", "vital signs"]
ch = 0

while ch != 4:
    
    print()
    print("1. Add to the list")
    print("2. Sort the list")
    print("3. Reverse the list")
    print("4. 999 to exit")
    ch = int(input("\nYour Selection: "))
    
    if ch == 1:
        print("\nCurrent list: ", end = "")
        for i in range(len(lis)):
            if i != (len(lis)-1):
                print(lis[i], end = ", ")
            else:
                print(lis[i])
        
        string = input("\nadd to list: ")
        lis.append(string)
        
        print("\nNew list: ", end = "")
        for i in range(len(lis)):
            if i != (len(lis)-1):
                print(lis[i], end = ", ")
            else:
                print(lis[i])
    elif ch == 2:
        lis = sorted(lis)
        print("\nSorted list: ", lis)
        
    elif ch == 3:
        lis.reverse()
        print("\nReverse list: ", lis)
    
    elif ch == 999:
        print("\n*** Good Bye ***")
        break
        
```
